```python
epocs = 50000

solucion_inicial_2 = instance.constructora2()

V = 10
rho = 0.3

n = len(low_levels)

h = 10000

P = np.ones((n, n))
T = np.zeros((n, n))

Q = np.ones((n, 2))
S = np.zeros((n, 2))

# inicializamos la matriz T_ij con P_ij / sum_{k in low_levels} P_ik
for i in range(n):
    suma_P = sum(P[i])
    suma_Q = sum(Q[i])
    T[i] = P[i] / suma_P
    S[i] = Q[i] / suma_Q
    
print(T)
print(S)

solucion_inicial = solucion_inicial_2 #Se comienza con una solución inicial s^c
mejor_solucion = solucion_inicial_2 #Se comienza con una mejor solución s*

secuencia = [] #Se inicia una secuencia vacía
matriz_secuencias = []

i_last = random.choice(low_levels)
print(i_last.id)

sum_epocs = 0


for count in range(epocs):
    costo_inicial = 1/solucion_inicial.objective_value + h*solucion_inicial.costo_infactible() if solucion_inicial.objective_value !=0 else 100000000000
    costo_mejor_solucion = 1/mejor_solucion.objective_value + h*mejor_solucion.costo_infactible() if mejor_solucion.objective_value !=0 else 100000000000
    id_i_next = seleccionar_segun_probabilidad(T[i_last.id-1]) #Setea la siguiente low_level
    i_next = low_levels[id_i_next]
    secuencia.append(id_i_next) #Agrega la siguiente low level a la lista
    u_next = random.choices([0,1], weights=S[i_next.id-1], k=1)[0] #Entrega u = 0 o 1 con probabilidad S_{i_next, u}
    
    if u_next == 1:
        solucion_temporal = i_next.implementacion(solucion_inicial) #Implementamos la low-level siguiente
        costo_temporal = 1/solucion_temporal.objective_value + h*solucion_temporal.costo_infactible() if solucion_temporal.objective_value !=0 else 100000000000
        if mejor_solucion.is_factible:
            rho = 10**-5 + V*(1-len(secuencia)/epocs)
        else:
            rho = 10**-3

        if costo_temporal < costo_inicial or costo_temporal < (1+rho)*costo_mejor_solucion:
            solucion_inicial = solucion_temporal
            #print(costo_temporal)
            costo_inicial = costo_temporal
             #Actualizar la matriz P y S
            Q[secuencia[-1], 1] =  Q[secuencia[-1], 1] + 1
            for i in range(len(secuencia)-1):
                i_k = secuencia[i]
                i_k_1 = secuencia[i+1]
                P[i_k, i_k_1] = P[i_k, i_k_1]+1
                Q[i_k, 0] = Q[i_k, 0]+1
                for i in range(n):
                    suma_P = sum(P[i])
                    suma_Q = sum(Q[i])
                    T[i] = P[i] / suma_P
                    S[i] = Q[i] / suma_Q
        if solucion_temporal.objective_value > mejor_solucion.objective_value:
            mejor_solucion = solucion_temporal
        
        secuencia = []
    print(mejor_solucion.objective_value, mejor_solucion.is_factible)
    i_last = i_next
    
print(mejor_solucion)
        
```
